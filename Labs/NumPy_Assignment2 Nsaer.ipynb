# Q3: Create a 3-D array in shape (2, 2, 3) containing the four arrays given below 
Note : the array's name is up to you.

|      |       |          |         |
|------| ----- | -------- | ------- |
|arr1 | 1.1    | 2.1       | 3.1        |
|arr2 | 4.1    | 5.1       | 6.1        |
|arr3 | 7.1    | 8.1       | 9.1        |
|arr4 | 10.1   | 11.1      | 12.1       |

```python
# write your code here ^_^

arr1 = [1.1, 2.1, 3.1]
arr2 = [4.1, 5.1, 6.1]
arr3 = [7.1, 8.1, 9.1]
arr4 = [10.1, 11.1, 12.1]

result = np.array([ [arr1, arr2], [arr3, arr4] ])

print(result)
```

# Q4: Print the following:
Note: use the same array from Q3.
- Array's type.
- Array's elements datatype.
- Array's shape.
- Array's size.
- Array's dimention.

```python
# write your code here ^_^
print("Array's type:", type(result))
print("Array's elements datatype:", result.dtype)
print("Array's shape:", result.shape)
print("Array's size:", result.size)
print("Array's dimension:", result.ndim)
```

# Q5:  Change the array dimention from 3-D to 4-D
Note: use the same array from Q3.

```python
# write your code here ^_^
new_array = result.reshape(1, 2, 2, 3)

print("New array's dimension:", new_array.ndim)
print("New array's shape:", new_array.shape)
```

# Q12: Join the given arrays below 
    arr1 = np.array([['A', 'B'], ['E', 'F']])
    arr2 = np.array([['C', 'D'], ['G', 'H']])
## Q12.1: Join the arrays without specifying the axis 

```python
# write your code here ^_^
arr1 = np.array([['A', 'B'], ['E', 'F']])
arr2 = np.array([['C', 'D'], ['G', 'H']])

joined_array = np.concatenate((arr1, arr2))

print(joined_array)
```

# Q12.2: Join the arrays along rows with axis = 1 

```python
# write your code here ^_^
joined_array = np.concatenate((arr1, arr2), axis=1)

print(joined_array)
```

# Q13: Split the array into two arrays with axis = 1, each array should contain four arrays. 
Note: use the same array from Q12.1

```python
# write your code here ^_^
split_arrays = np.hsplit(joined_array, 2)


print("First split array:")
print(split_arrays[0])

print("\nSecond split array:")
print(split_arrays[1])
```
